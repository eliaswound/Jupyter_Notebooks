```python
#Testing Expanding labels. 
import os
import sys
# import tk for getting the directory faster. dont need this in a command line/server version
import tkinter as tk
from tkinter import filedialog
# you need numpy as always
import numpy as np
import glob
import tifffile 
from tifffile import imread
from skimage import segmentation
import math
import bigfish.multistack as multistack
```

```python
# Load your labels
def open_file_dialog():
    root = tk.Tk()
    root.withdraw()  # Hide the main window

    file_path = filedialog.askdirectory()  # Open the file dialog window
    os.chdir(file_path)
    root.destroy()  # Destroy the root window after file selection
    return file_path

if __name__ == "__main__":
    selected_file_path = open_file_dialog()
    print("Selected file path:", selected_file_path)
```

```python
filenames = glob.glob('*000.tif')
dataset = list(map(imread,filenames))
```

```python
expanded_labels = segmentation.expand_labels(dataset[0], distance=20)
tifffile.imwrite('20expandedlabel.tif', expanded_labels, photometric = 'minisblack')
```

```python
# Load your coordinates
def open_file_dialog():
    root = tk.Tk()
    root.withdraw()  # Hide the main window

    file_path = filedialog.askdirectory()  # Open the file dialog window
    os.chdir(file_path)
    root.destroy()  # Destroy the root window after file selection
    return file_path

if __name__ == "__main__":
    selected_file_path = open_file_dialog()
    print("Selected file path:", selected_file_path)
```

```python
filenames_coord = glob.glob('*cluster.npy')
coord = np.load(filenames_coord[0])
```

```python
coord[:, 0] //= 15
new_array = np.array([row[:3] for row in coord],dtype =np.int32)
```

```python
cell_coord_results = multistack.extract_cell(cell_label = expanded_labels,ndim =3,nuc_label=dataset[0],rna_coord = new_array)
```

```python
for i, cell_results in enumerate(cell_coord_results):
    print("cell {0}".format(i))
    
    # get cell results
    cell_mask = cell_results["cell_mask"]
    cell_coord = cell_results["cell_coord"]
    nuc_mask = cell_results["nuc_mask"]
    nuc_coord = cell_results["nuc_coord"]
    rna_coord = cell_results["rna_coord"]

    print("\r number of rna {0}".format(len(rna_coord)))

```
